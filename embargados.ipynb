```python
df_embargos = ps.sqldf(
    '''
    SELECT * 
    FROM df_productos
    INNER JOIN df_embargos
    ON df_productos.TIPO_IDENTIFICACION_TITULAR = df_embargos.DOCTYPE_NUM
    AND df_productos.NRO_DE_IDENTIFICACION_DEL_TITULAR = df_embargos.RESPONDENTID
    '''
, locals())
df_embargos.shape
```

```python
df_embargos.columns
```

```python
def cast_doctype_num_to_name(row):
    if row["DOCTYPE_NUM"] == "1" :
        return "Cedula de Ciudadanía"
    if row["DOCTYPE"] == "2":
        return "NIT"
    return "Otros"
```

```python
df_embargos["tipo_documento"] = df_embargos.apply(cast_doctype_num_to_name, axis=1)
df_embargos[["tipo_documento", "DOCTYPE_NUM"]]
```

```python
df_embargos[["NRO_DE_IDENTIFICACION_DEL_TITULAR"]]
type(df_embargos[["NRO_DE_IDENTIFICACION_DEL_TITULAR"]])
pd.unique(df_embargos["NRO_DE_IDENTIFICACION_DEL_TITULAR"])
```

```python
df_embargos["numero_documento"] = df_embargos["NRO_DE_IDENTIFICACION_DEL_TITULAR"]
df_embargos["FECHA_DE_APERTURA_DE_LA_CUENTA"].head()
```

```python
def format_date(row):
    date = row["FECHA_DE_APERTURA_DE_LA_CUENTA"]
    if len(date) == 8:
        year = date[:4]
        month = date[4:6]
        day = date[6:]
        concatenated_date = f"{year}-{month}-{day}"
        return str(datetime.strptime(concatenated_date, "%Y-%m-%d").date())
    return "2000-02-02"
```

```python
df_embargos["fecha_apertura"] = df_embargos.apply(format_date, axis=1)
```

```python
def check_status(row):
    status = str(row["MARCAS_DE_LAS_CUENTAS"])
    if status[21] == "Y":
        return "E"
    if status[22] == "Y":
        return "N"
    if status[23] == "Y":
        return "F o S"
    return "Otros"
```

```python
df_embargos["MARCAS_DE_LAS_CUENTAS"]
```

```python
df_embargos["estado"] = df_embargos.apply(check_status,axis=1)
```

```python
df_embargos["estado"].head(30)
```

```python
def check_plan(row):
    status = str(row["MARCAS_DE_LAS_CUENTAS"])
    if status[10] == "Y":
        return "Ahorro a la mano"
    if status[11] == "Y":
        return "AFC"
    if status[12] == "Y":
        return "Empleado"
    if status[13] == "Y":
        return "Nomina"
    return "Otros"
```

```python
df_embargos["plan"] = df_embargos.apply(check_plan,axis=1)
```

```python
def format_saldo_disponible(row):
    if len(row["SALDO_EFECTIVO_DE_LA_CUENTA"]) == 20:
        left = str(int(str(row["SALDO_EFECTIVO_DE_LA_CUENTA"])[:16]))
        right = (str(row["SALDO_EFECTIVO_DE_LA_CUENTA"])[-4:])[:2]
        return f"{left}.{right}"
```

```python
df_embargos["saldo_disponible_tmp"] = df_embargos.apply(format_saldo_disponible, axis=1)
df_embargos["saldo_disponible_tmp"]
```

```python
df_embargos["saldo_disponible"] = df_embargos["SIGNO_DEL_SALDO_EFECTIVO_DE_LA_CUENTA"] + df_embargos["saldo_disponible_tmp"]
```

```python
df_embargos[["SIGNO_DEL_SALDO_EFECTIVO_DE_LA_CUENTA", "SALDO_EFECTIVO_DE_LA_CUENTA", "saldo_disponible"]]
```

```python
def check_plan(row):
    status = str(row["MARCAS_DE_LAS_CUENTAS"])
    if status[3] and status[14] == "N":
        return "SI"
    return "NO"
```

```python
df_embargos["cuenta_inembargable"] = df_embargos.apply(check_plan, axis=1)
```

```python
df_embargos["nombre_completo"] = df_embargos["NOMBRE_TITULAR_CUENTA"].apply(lambda x: x.upper())
```

```python
df_embargos["nombre_corto"] = df_embargos["NOMBRE_TITULAR_CUENTA"].apply(lambda x: x.split(" ")[0])
```

```python
# df_embargos["oficina_duena"] = df_embargos["OFICINA_DUENA_DE_LA_CUENTA"].apply(lambda x: x.upper())
```

```python
def check_marcacion_cuenta_exenta_gmf(row):
    status = str(row["MARCAS_DE_LAS_CUENTAS"])
    if status[19] == "N" or status[19] == "T":
        return "T"
    return "N"
```

```python
df_embargos["marcacion_cuenta_exenta_gmf"] = df_embargos.apply(check_marcacion_cuenta_exenta_gmf, axis=1)
```

```python
df_embargos["cuenta_embargada"] = "SI"
```
