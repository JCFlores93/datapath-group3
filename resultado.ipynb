```python
# resultados - regla 3
df_embargos["CITY"] = df_embargos["CITY"].astype(str)
df_embargos["CLAIMANTENTITY"] = df_embargos["CLAIMANTENTITY"].astype(str)

df_embargos["CLAIMANTENTITY"].fillna(value = '', inplace = True)
```

```python

def check_localidad(row):
    
    if row['CLAIMANTENTITY']!='nan' and len(row['CLAIMANTENTITY'])!=0:
        return row['CITY'] + " (" + row['CLAIMANTENTITY'] + ")"
    if row['CLAIMANTENTITY']=='nan' or len(row['CLAIMANTENTITY'])==0:
        return row['CITY']
    if len(row['CLAIMANTENTITY'])==0:
        return "sin localidad"

```

```python
df_embargos["localidad"] = df_embargos.apply(check_localidad, axis = 1)
```

```python
def validar_fecha(row):
    
    date = row["RECEPTIONDATE"]
    if len(date) == 16:
        year = date[6:10]
        month = date[3:5]
        day = date[0:2]
        
        hour = date[11:13]
        minute = date[14:16]
    
        concatenated_date = f"{year}-{month}-{day}"
        return str(datetime.strptime(concatenated_date, "%Y-%m-%d").date())
     
# "%h:%m"
# str(datetime.strptime(concatenated_date, "%H:%M").time())
```

```python
df_embargos["fecha_recepcion"] = df_embargos.apply(validar_fecha, axis = 1)
```

```python
# resultados - regla 8

def validar_hora(row):
    
    date = row["RECEPTIONDATE"]
    if len(date) == 16:
        
        hour = date[11:13]
        minute = date[14:16]
        concatenated_date = f"{hour}:{minute}"
        return str(datetime.strptime(concatenated_date, "%H:%M").time())
    

     
# "%h:%m"
# str(datetime.strptime(concatenated_date, "%H:%M").time())
```

```python
df_embargos["hora_recepcion"] = df_embargos.apply(validar_hora, axis = 1)
```
